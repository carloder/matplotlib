```

df3['Time']=df3.Time.str.split(expand=True,)[0]
df333=df3.groupby(['Point ID', 'Time'], as_index=False).agg({'Target Easting [m]':'mean','Target Northing [m]': 'mean','Target Elevation [m]': 'mean'})
Td=df333.Time.str.split("-",expand=True,)[0]
Tm=df333.Time.str.split("-",expand=True,)[1]
Ta=df333.Time.str.split("-",expand=True,)[2]
df333['Td']=df333.Time.str.split("-",expand=True,)[0]
df333['Tm']=df333.Time.str.split("-",expand=True,)[1]
df333['Ta']=df333.Time.str.split("-",expand=True,)[2]
df333.sort_values(by=['Point ID','Ta','Tm','Td'], inplace=True)
df333
h1=set(df3['Point ID'])
h1s=sorted(list(h1))
end=len(h1s)
df2i=df333[df333['Point ID'] == h1s[0]]
df2i=df2i.reset_index(drop=True)
df2i['DHA']=((df2i['Target Easting [m]']-df2i['Target Easting [m]'][0])**2+(df2i['Target Northing [m]']-df2i['Target Northing [m]'][0])**2)**0.5
for i in  range(1,end):
    df2=df333[df333['Point ID'] == h1s[i]]
    df2=df2.reset_index(drop=True)
    df2['DHA']=((df2['Target Easting [m]']-df2['Target Easting [m]'][0])**2+(df2['Target Northing [m]']-df2['Target Northing [m]'][0])**2)**0.5
    df2i=df2i.append(df2)
df333=df2i
df333
df333.rename({'Target Northing [m]': 'N'}, axis=1, inplace=True)
df333.rename({'Target Elevation [m]': 'Elv'}, axis=1, inplace=True)
df333.rename({'Target Easting [m]': 'EE'}, axis=1, inplace=True)

import altair as alt
alt.renderers.enable('notebook')
df333.dtypes 
df333['Time'] = pd.to_datetime(df333['Time'])
df333['T11'] = df333['Time'].dt.strftime('%d/%m/%y') 
df333['T11'] = pd.to_datetime(df333['T11'])
alt.data_transformers.enable('default', max_rows=None)
alt.Chart(df333, width=1200, height=600).mark_line().encode(
    alt.X('T11', scale=alt.Scale(zero=False)),
    alt.Y('DHA', scale=alt.Scale(zero=False)),
    color='Point ID',
    tooltip=['Point ID','DHA','T11'],
).interactive()
```

# **3**

```
h1=set(df3['Point ID'])
h1s=sorted(list(h1))
end=len(h1s)
for k in range(0,end):
    df22=df333[df333['Point ID'] == h1s[k]]
    df22=df22.reset_index(drop=True)
    df22['EE2']=df22['EE']*0
    df22['N2']=df22['N']*0
    df22['Elv2']=df22['Elv']*0
    df22

    end=len(df22)
    for i in  range(0,end-1):
        df22['EE2'].iloc[i+1]=df22['EE'].iloc[i]-df22['EE'].iloc[i+1]
        df22['N2'].iloc[i+1]=df22['N'].iloc[i]-df22['N'].iloc[i+1]
        df22['Elv2'].iloc[i+1]=df22['Elv'].iloc[i]-df22['Elv'].iloc[i+1]
        df22
        if abs(df22['EE2'].iloc[i+1])>0.0050:
           for j in range (i, end-1):
               df22['EE'].iloc[j+1]=df22['EE'].iloc[j+1]+df22['EE2'].iloc[i+1]
        if abs(df22['N2'].iloc[i+1])>0.0050:
           for j in range (i, end-1):
               df22['N'].iloc[j+1]=df22['N'].iloc[j+1]+df22['N2'].iloc[i+1]
        if abs(df22['Elv2'].iloc[i+1])>0.0050:
           for j in range (i, end-1):
               df22['Elv'].iloc[j+1]=df22['Elv'].iloc[j+1]+df22['Elv2'].iloc[i+1]
    df22
    if k == 0:
       df2join=df22
    else:
       df2join=df2join.append(df22)
h1=set(df2join['Point ID'])
h1s=sorted(list(h1))
end=len(h1s)
df2i=df2join[df2join['Point ID'] == h1s[0]]
df2i=df2i.reset_index(drop=True)
df2i['DHA']=((df2i['EE']-df2i['EE'][0])**2+(df2i['N']-df2i['N'][0])**2)**0.5
for i in  range(1,end):
    df2=df2join[df2join['Point ID'] == h1s[i]]
    df2['DHA']=((df2['EE']-df2['EE'][0])**2+(df2['N']-df2['N'][0])**2)**0.5
    
    df2i=df2i.append(df2)
df2join=df2i
df2join

import altair as alt
df2join.dtypes 
df2join['Time'] = pd.to_datetime(df2join['Time'])
df2join['T11'] = df2join['Time'].dt.strftime('%d/%m/%y')
df2join['T11'] = pd.to_datetime(df2join['T11'])
alt.data_transformers.enable('default', max_rows=None)
alt.Chart(df2join, width=1200, height=600).mark_line().encode(
    alt.X('T11', scale=alt.Scale(zero=False)),
    alt.Y('DHA', scale=alt.Scale(zero=False)),
    color='Point ID',
    tooltip=['Point ID','DHA','T11'],
).interactive()
```

```
#####velocidad
h1=set(df2join['Point ID'])
h1s=sorted(list(h1))
end=len(h1s)
for j in  range(0,end):
    d2=df2join[df2join['Point ID'] == h1s[j]]
    d2=d2.reset_index(drop=True)
    d2['vel']=d2['DHA']*0
    end1=len(d2)
    for i in  range(0,end1-1):
        d2['vel'].iloc[i+1]=(d2['DHA'].iloc[i+1]-d2['DHA'].iloc[i])*100/(d2['T11'].iloc[i+1]-d2['T11'].iloc[i]).days
        d2
    if j == 0:
       d2join=d2
    else:
       d2join=d2join.append(d2)
d2join
         
```

# **8 tabla para el informeee**

```
import numpy as np
h1=set(df2join['Point ID'])
h1s=sorted(list(h1))
end=len(h1s)
hjj=d2join['Point ID'].drop_duplicates()
list_of_tuples = list(zip(hjj)) 
list_of_tuples  
hjj = pd.DataFrame(list_of_tuples, columns = ['id']) 
hjj['DHAmax']=np.nan
hjj['vel']=np.nan
hjj['FECHA_inicial']=np.nan
hjj['FECHA_final']=np.nan
hjj['deltaD']=np.nan
hjj['frecuencia']=np.nan
for i in range (0,end):
    hj=d2join[(d2join['Point ID'] == h1s[i])&(d2join['Ta'] == "19")&(d2join['Tm'] == "05")]
    hj=hj.reset_index(drop=True)
    hjw=d2join[(d2join['Point ID'] == h1s[i])&(d2join['Ta'] == "19")]
    hjw=hjw.reset_index(drop=True)
    end1=len(hj['Point ID'])
    for j in range (0,end1):
        mdmens=hj['DHA'].max()
        mvmens=hj['vel'].max()
        hjj['DHAmax'].iloc[i]=mdmens
        hjj['vel'].iloc[i]=mvmens
hjj
for i in range (0,end):
    hjw=d2join[(d2join['Point ID'] == h1s[i])&(d2join['Ta'] == "19")]
    hjw=hjw.reset_index(drop=True)
    endw=len(hjw['Point ID'])
    for j in range (0,endw):
        hjj['FECHA_inicial'].iloc[i]=hjw['T11'].iloc[0]
        hjj['FECHA_final'].iloc[i]=hjw['T11'].iloc[endw-1]
        hjj['deltaD'].iloc[i]= hjj['FECHA_final'].iloc[i]-hjj['FECHA_inicial'].iloc[i]
        hjj['frecuencia'].iloc[i]=endw
hjj
```
